#Step 5: Visualize the Dataset

In this step, we will create our first graphs using matplotlib.

- Visualization helps us understand data patterns, trends, and relationships without reading large tables.
- We need to import our visualizor tool in python which is matplotlib

```python
import matplotlib.pyplot as plt  # matplotlib is a library to make plots in Python
```

We are plotting each company as a point on a graph, where the x-position shows their golf handicap and the y-position shows how well the company performs. Then we label the axes and give the plot a title so it‚Äôs clear what the viewer is looking at.

What this code does:

- plt.scatter(df['Handicap'], df['Performance']) ---> This line draws the dots on the graph.

  - X‚Äëaxis ‚Üí df['Handicap'] (the golf handicap values for each company)

  - Y‚Äëaxis ‚Üí df['Performance'] (the corporate performance values)

- plt.xlabel('Golf Handicap') ---> Labeling axes is important because without labels, the viewer will not understand the chart.


- plt.ylabel('Corporate Performance') ---> This names the y‚Äëaxis so we know what the vertical numbers mean.


- plt.title('Golf vs Corporate Performance') -- > Adds a title at the top of the graph. A good title summarizes the story the graph is trying to tell.


- plt.show() --> This actually displays the plot in the notebook. Without show(), the graph might not appear.

Think of it as saying, ‚ÄúOkay, Python, please show my picture now!‚Äù

```python
# Assuming your DataFrame is df
# Scatter plot: Golf Handicap vs Corporate Performance
plt.scatter(df['Handicap'], df['Performance'])

# Add labels and title
plt.xlabel('Golf Handicap')
plt.ylabel('Corporate Performance')
plt.title('Golf vs Corporate Performance')

plt.show()
```

This scatter plot shows the relationship between a company‚Äôs golf handicap and its corporate performance. Each dot on the graph represents a single company. The x-axis measures golf handicap, where lower numbers mean better golf skill, and the y-axis measures the company‚Äôs performance. By plotting the data this way, we can visually explore whether there is a pattern: for example, do CEOs with lower golf handicaps tend to lead companies with higher performance? Visualizations like this help us quickly identify trends, outliers, and relationships in the data that are harder to see in a raw table of numbers.

Step by step, how can wee add name for each data point on the scatter plot.

for i, txt in enumerate(df['Company']):

This goes through the list of company names one by one.

i is the number of the row (0, 1, 2, ‚Ä¶).

txt is the company name (like "Apple" or "Google").

plt.annotate(txt, (df['Handicap'][i], df['Performance'][i]))

annotate means write text on the graph.

txt is the name we want to write.

(df['Handicap'][i], df['Performance'][i]) is the location of the dot for that company.

fontsize=8 makes the name small.

alpha=0.7 makes the name a bit transparent so it doesn‚Äôt clutter the chart.

```python
# Assuming your DataFrame is df
# Scatter plot: Golf Handicap vs Corporate Performance
plt.figure(figsize=(8, 5))
plt.scatter(df['Handicap'], df['Performance'], color='green')

# Add labels and title
plt.xlabel('Golf Handicap (Lower is Better)')
plt.ylabel('Corporate Performance')
plt.title('Golf Handicap vs Corporate Performance')

# Optional: show company names next to points
for i, txt in enumerate(df['Company']):
    plt.annotate(txt, (df['Handicap'][i], df['Performance'][i]), fontsize=8, alpha=0.7)

plt.grid(True)
plt.show()
```
