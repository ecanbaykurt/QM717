##Step 1: Import your Python Libraries

```python
# Pandas is for handling tables (like Excel)
import pandas as pd

# NumPy is for numbers and arrays
import numpy as np

# Matplotlib is for charts and graphs
import matplotlib.pyplot as plt
```

##Step 2: Load the S&P 500 Companies Data

```python
# Replace the filename with your actual CSV or Excel file name

sp500_df = pd.read_excel('/content/S&P 500 constituents.xlsx')  # Example name
sp500_df.head()  # Shows the first few rows so you know it worked!
```

##Step 3: Frequency Distribution of Sectors

```python
# Count how many companies are in each sector

sector_counts = sp500_df['GICS Sector'].value_counts()
print("Frequency Table:\n", sector_counts)

# Relative Frequency = what percent each sector is
sector_relative = sector_counts / sector_counts.sum()
print("\nRelative Frequency Table:\n", sector_relative)
```

```python
# Bar chart: Frequency of each sector

sector_counts.plot(kind='bar', figsize=(10, 5), color='skyblue', edgecolor='black')
plt.title("Frequency of S&P 500 Companies by Sector")
plt.xlabel("Sector")
plt.ylabel("Number of Companies")
plt.xticks(rotation=45)
plt.show()
```

##Step 4: Frequency of Subsectors in Your Chosen Sector

```python
# Pick a sector youâ€™re interested in
chosen_sector = 'Information Technology'  # You choose!

# Filter only that sector
subsector_df = sp500_df[sp500_df['GICS Sector'] == chosen_sector]

# Count subsectors
subsector_counts = subsector_df['GICS Sub-Industry'].value_counts()
subsector_relative = subsector_counts / subsector_counts.sum()

print(f"Subsectors in {chosen_sector}:\n", subsector_counts)
print("\nRelative Frequencies:\n", subsector_relative)
```

```python
# Bar chart: Subsector breakdown

subsector_counts.plot(kind='bar', figsize=(8, 5), color='salmon', edgecolor='black')
plt.title(f"{chosen_sector} Subsectors")
plt.xlabel("Subsector")
plt.ylabel("Number of Companies")
plt.xticks(rotation=45)
plt.show()
```
